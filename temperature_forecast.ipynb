```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = pd.read_csv("temperature.csv")

temperature = data.AverageTemperatureFahr
days = data.Days
```

```python
figure = plt.plot(days, temperature, 'ro', ms=10, mec='k')
plt.ylabel("Temperature")
plt.xlabel("Days")
```

```python
clt = LinearRegression()
days = days.values.reshape(-1,1)
clt.fit(days, temperature)
prediction = clt.predict(days)
```

```python
plt.scatter(days, temperature)
plt.plot(days, prediction)
```
